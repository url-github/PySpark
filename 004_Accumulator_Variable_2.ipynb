```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

`SparkContext`, jest głównym obiektem kontrolującym operacje na klastrze Spark. Działa jako “serce” klastra Spark i zarządza wykonywaniem zadań.

```python
sc = spark.sparkContext
spark
```

```python
game = [('USA', 'Poland', 9),
       ('Germany','Italy', 5),
       ('Italy','India', 2),
       ('India', 'Italy', 4),
       ('France','Poland', 1),
       ('Italy','India', 2),
       ('USA','France', 3),
       ('India','France', 7)]
```

```python
df = spark.createDataFrame(game, schema=["home_country","host_country","matches"])
df.printSchema()
df.show()
```

```python
matchcounter = sc.accumulator(0)
matchcounter
```

```python
def counterfn(match_num):
    home = match_num["home_country"]
    host = match_num["host_country"]
    if home == 'Poland' or host == 'Poland':
        matchcounter.add(match_num["matches"])
```

```python
df.foreach(lambda match_num : counterfn(match_num))
```
