```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
dataset = [("Bob", 42, 'IT', 2000),
           ("Lisa", 59,  'Finance', 3000),
           ("Diane", 20,  'IT', 5000),
           ("Janet", 60,  'Finance', 7000),
           ("Calvin", 26, 'IT', 4000),
           ("Elena", 24,  'Finance', 6000),
           ("Bill", 50,  'IT', 1000),
           ("Dawn", 33,  'Finance', 8000),
            ("Ken", 66,  'IT',4500),
           ("Lee", 21, 'Finance',6500),
           ("Asha", 53, 'IT',1500),
          ("Marsh", 34,'Finance', 8500)]

col = ['first_name', 'age', 'deptartment', 'salary']

employeedf = spark.createDataFrame(data = dataset,schema = col)
```

```python
employeedf.show()
```

```python
employeedf.write.option("header",True) \
        .partitionBy("deptartment") \
        .mode("overwrite") \
        .csv("employee")
```

```python
emp = [("Bob", 42, 'Sales', 2000),
           ("Lisa", 59,  'Finance', 3000),
           ("Diane", 20,  'IT', 5000),
           ("Janet", 60,  'Finance', 7000),
           ("Calvin", 26, 'IT', 4000),
           ("Elena", 24,  'Sales', 6000),
           ("Bill", 50,  'IT', 1000),
           ("Dawn", 33,  'Finance', 8000),
            ("Ken", 66,  'IT',4500),
           ("Lee", 21, 'Sales',6500),
           ("Asha", 53, 'IT',1500),
          ("Marsh", 34,'Finance', 8500)]

col = ['first_name', 'age', 'deptartment', 'salary']

empdf = spark.createDataFrame(data = emp, schema = col)
```

```python
empdf.show()
```

`coalesce(2)` służy do zapisania DataFrame empdf w formacie CSV z ograniczeniem liczby plików wyjściowych do 2.

```python
empdf.coalesce(2) \
        .write.option("header",True) \
        .mode("overwrite") \
        .csv("emp")
```

`repartition(2)` pozwala na pełne przetasowanie danych, co oznacza, że Spark dzieli dane na określoną liczbę partycji niezależnie od istniejącej liczby partycji.

```python
empdf.repartition(2) \
        .write.option("header",True) \
        .mode("overwrite") \
        .csv("emp1")
```
